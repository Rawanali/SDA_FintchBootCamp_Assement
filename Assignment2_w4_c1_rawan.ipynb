```python
import pandas as pd
import numpy as np
```

```python
# Parameters:  Values

growth_factor = 1.18 # 12% of prior period Revenue
EBITDA_margin = 0.45 # of current Revenue
Taxes= 0.21 # of current EDBITDA
capital_Exp = 0.03 # of current Revenue
work_capital = 0.06 # of Revenue
exit_multi = 32.70
disc_fac = 0.94 # (6%). 
discount_rate =0.06
```

```python
dates = pd.date_range(start='2022',end='2031',freq='Y')
```

```python
modle_df = pd.DataFrame(index= dates)
last_value_df = pd.DataFrame()
```

```python
# Revenue Growth 
initial_revenue = 222

rev_pro = [initial_revenue * growth_factor] #+ rev_start_v, list

for i in range(1,len(dates)):
    rev_pro.append(rev_pro[i-1]*growth_factor)

modle_df['Revenu'] = rev_pro
modle_df['Revenu'] = modle_df['Revenu'].round(2)
```

```python
modle_df
```

```python
# EBITDA 
    
modle_df['EBITDA']= modle_df['Revenu']*  EBITDA_margin 
modle_df['EBITDA']= modle_df['EBITDA'].round(2)
```

```python
modle_df
```

```python
# Taxis 
    
modle_df['Taxis']= modle_df['EBITDA']*  -Taxes
modle_df['Taxis']= modle_df['Taxis'].round(2)
```

```python
modle_df
```

```python
# Capital Expenditures

modle_df['Cap Exp']= modle_df['Revenu']*  -capital_Exp 
modle_df['Cap Exp'] = modle_df['Cap Exp'].round(2)
```

```python
modle_df
```

```python
# Working Capital
modle_df['Work_cap']= modle_df['Revenu']* -work_capital
modle_df['Work_cap'] = modle_df['Work_cap'].round(2)
```

```python
modle_df
```

```python
# Cash Flow
modle_df['Cash_flow']=modle_df.drop(['Revenu'],axis=1).sum(axis=1).round(2)
```

```python
modle_df
```

```python
# Discount Factor
discout_factor = [disc_fac **i for i in range(1,len(dates)+1)]
modle_df['discout_factor'] = discout_factor
modle_df['discout_factor'] = modle_df['discout_factor'].round(2)
```

```python
modle_df
```

```python

modle_df['present_value_fcf'] = discout_factor * modle_df['Cash_flow'].values
modle_df['present_value_fcf'] = modle_df['present_value_fcf'].round(2)
```

```python
modle_df.transpose()
```

# End of frist Data Frame 
      
      

```python
last_value_df = pd.DataFrame(index= ['2030'])
```

```python
last_value_df['Revenu'] = round(modle_df['Revenu'].iloc[-1]*growth_factor)
last_value_df['EBITDA'] = last_value_df['Revenu']*EBITDA_margin
last_value_df['prisent_value_of_projected'] = modle_df['present_value_fcf'].sum().round(2)
last_value_df['present_value_of_asset']=round(modle_df.iloc[-1]['Revenu'] 
                                              * 1.12 * 0.50 * 10.0 * modle_df.iloc[-1]['discout_factor'])
```
